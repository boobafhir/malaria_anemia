<h4>1 Artificial Neural Network</h4>

```python
from keras.models import Sequential
from keras.layers import Dense
import numpy
import csv as csv
import pandas as pd
df = pd.read_csv('sdhmalaria_redu.csv')
df=df.dropna()
shape=df.shape
print(shape)
array=df.values
X = array[:,0:27]
Y = array[:,27]

# create model
model = Sequential()
model.add(Dense(20, input_dim=27 , activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
# Compile model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X,Y, epochs=150, batch_size=10)
scores = model.evaluate(X, Y)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```

<h4>2. Random Forests </h4>

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
Ym=df.Malaria_test
Xm=df.drop('Malaria_test', axis=1)

X_train, X_test, y_train, y_test = train_test_split(Xm, Ym,test_size=0.2,random_state=1, stratify=Ym)

random_forest = RandomForestClassifier(n_estimators=100, max_depth=10, random_state=50)
random_forest.fit(X_train,y_train)

predictionRF= random_forest.predict(X_test)
print("Accuracy:",accuracy_score(y_test, predictionRF) * 100)
```

<h4>3.KNN</h4>

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train,y_train)
predictionKNN=knn.predict(X_test)
print("Accuracy:",accuracy_score(y_test,predictionKNN)*100)
```

<h4>4.SVM</h4>

```python
from sklearn import svm
model = svm.SVC(kernel='linear', C=1.0) 
model.fit(Xm, Ym)
model.score(Xm, Ym)
predicted= model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, predicted)*100)
```

<h4>5. Naive Bayes</h4>

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
model2=GaussianNB()
model2.fit(Xm,Ym)
target_pred=model2.predict(X_test)
print('Accuracy:',accuracy_score(y_test, target_pred) * 100)
```
